### Preprocessing Discrete Variables: Automating Calculations

```python
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    '''
    @df: The dataframe containing the data
    @discrete_variable_name: The name of the discrete variable of interest
    @good_bad_variable_df: Good/Bad column data 
    '''

    # concatenate the discrete variable and the good/bad column
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis=1) 
    # Add the count and meand per discrete variable value
    df = pd.concat([df.groupby(df.columns.values[0], as_index= False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index= False)[df.columns.values[1]].mean()], axis=1)
    # delete the redundant first variable
    df = df.iloc[:, [0, 1, 3]]
    # Rename the column
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    # Compute the proportion of observations
    df['prop_n_obs'] = df['n_obs']/df['n_obs'].sum()
    # Compute the number of Good/Bad borrower
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1-df['prop_good'])* df['n_obs']
    # Compute the proportion of Godd/Bad borrower from the total number of borrowers
    df['prop_n_good'] = df['n_good']/df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad']/df['n_bad'].sum()
    # Compute the Weight of Evidence
    df['WoE'] = np.log(df['prop_n_good']/df['prop_n_bad'])
    # Sort by Weight of Evidence and reset the index
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop= True)
    # Compute the differences in proportion of good loans and weight evidence 
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    # Find the information value
    df['IV'] = (df['prop_n_good']-df['prop_n_bad'])* df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

```python
def plot_by_woe(df_woe, rotation_of_x_axis_labels=0):
    x = np.array(df_woe.iloc[:, 0].apply(str))
    y = df_woe['WoE']
    plt.figure(figsize=(18, 6))
    plt.plot(x,y, marker = 'o', linestyle= '--', color = 'k')
    plt.xlabel(df_woe.columns[0])
    plt.ylabel('weight of Evidence')
    plt.title(str('Weight of Evidence by ' + df_woe.columns[0] ))
    plt.xticks(rotation = rotation_of_x_axis_labels)
    
```

```python
# We can bind 2gether the home_ownership category that have small population (underrepresented categories)
# We have to create a dummy variables for home_ownership (OTHER, NONE, RENT, ANY) and another dummy for MORTGAGE and another for OWN.
# These choice are based on the proportion of borrowers, good and the WoE.

# There are many categories with very few observations and many categories with very different "good" %.
# Therefore, we create a new discrete variable where we combine some of the categories.
# 'OTHERS' and 'NONE' are riskiest but are very few. 'RENT' is the next riskiest.
# 'ANY' are least risky but are too few. Conceptually, they belong to the same category. Also, their inclusion would not change anything.
# We combine them in one category, 'RENT_OTHER_NONE_ANY'.
# We end up with 3 categories: 'RENT_OTHER_NONE_ANY', 'OWN', 'MORTGAGE'.

df_inputs_prepr['home_ownership:RENT_OTHER_NONE_ANY'] = sum([df_inputs_prepr['home_ownership:RENT'], df_inputs_prepr['home_ownership:OTHER'],
                                                             df_inputs_prepr['home_ownership:NONE'], df_inputs_prepr['home_ownership:ANY']])

# 'RENT_OTHER_NONE_ANY' will be the reference category.

# Alternatively:
#loan_data.loc['home_ownership' in ['RENT', 'OTHER', 'NONE', 'ANY'], 'home_ownership:RENT_OTHER_NONE_ANY'] = 1
#loan_data.loc['home_ownership' not in ['RENT', 'OTHER', 'NONE', 'ANY'], 'home_ownership:RENT_OTHER_NONE_ANY'] = 0
#loan_data.loc['loan_status' in ['OWN'], 'home_ownership:OWN'] = 1
#loan_data.loc['loan_status' not in ['OWN'], 'home_ownership:OWN'] = 0
#loan_data.loc['loan_status' in ['MORTGAGE'], 'home_ownership:MORTGAGE'] = 1
#loan_data.loc['loan_status' not in ['MORTGAGE'], 'home_ownership:MORTGAGE'] = 0
```

### Preprocessing Continuous Variables

```python
def woe_ordered_continuous(df, continuous_variable_name, good_bad_variable_df):
    '''
    @df: The dataframe containing the data
    @continuous_variable_name: The name of the discrete variable of interest
    @good_bad_variable_df: Good/Bad column data 

    Here we've to sort the data with respect to the value
    '''

    # concatenate the discrete variable and the good/bad column
    df = pd.concat([df[continuous_variable_name], good_bad_variable_df], axis=1) 
    # Add the count and meand per discrete variable value
    df = pd.concat([df.groupby(df.columns.values[0], as_index= False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index= False)[df.columns.values[1]].mean()], axis=1)
    # delete the redundant first variable
    df = df.iloc[:, [0, 1, 3]]
    # Rename the column
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    # Compute the proportion of observations
    df['prop_n_obs'] = df['n_obs']/df['n_obs'].sum()
    # Compute the number of Good/Bad borrower
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1-df['prop_good'])* df['n_obs']
    # Compute the proportion of Godd/Bad borrower from the total number of borrowers
    df['prop_n_good'] = df['n_good']/df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad']/df['n_bad'].sum()
    # Compute the Weight of Evidence
    df['WoE'] = np.log(df['prop_n_good']/df['prop_n_bad'])
    # Sort by Weight of Evidence and reset the index
    # df = df.sort_values(['WoE'])
    # df = df.reset_index(drop= True)
    # Compute the differences in proportion of good loans and weight evidence 
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    # Find the information value
    df['IV'] = (df['prop_n_good']-df['prop_n_bad'])* df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```
